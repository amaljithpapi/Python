```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import mean_absolute_error
```

```python
data = pd.read_csv('/content/winequality-red.csv')
data
```

```python
corrmat = data.corr()
plt.subplots(figsize = (12,9))
sns.heatmap(corrmat, annot = True,vmin= -0.2, vmax = 0.9, cmap = 'YlGnBu')
plt.show()
```

```python
data.columns
```

```python
data.head()
```

```python
x = data.drop('quality', axis=1)
y = data['quality']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_test
```

```python
y_test
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
pred = model.predict(x_test)
mse = mean_squared_error(y_test, pred)
r2 = r2_score(y_test, pred)
```

```python
mse, r2
```

```python
mse = mean_absolute_error(y_test, pred)
print(mse)
```

```python
plt.scatter(x['alcohol'], y, color = 'red', label='Actual Quality')
plt.plot(x['alcohol'], model.predict(x), color='blue', label='Predicted Quality')

plt.title('Linear Regression: Alcohol vs Wine Quality')
plt.xlabel('Alcohol Content')
plt.ylabel('Wine Quality')
plt.show()
```
